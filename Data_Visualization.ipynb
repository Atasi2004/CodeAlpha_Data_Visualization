# Data Visualization on FIFA World Cup Players Dataset

## Objective

The objective of this project is to visualize and analyze the FIFA World Cup Players dataset using Python libraries such as Pandas, Matplotlib, and Seaborn. The project aims to transform raw data into meaningful visualizations, identify patterns and trends, and present insights that help in understanding player characteristics and team distributions.

# 📈 Visualization 1
Top 10 Countries with the Most Players records

```python
plt.figure(figsize=(10,6))
df['Team Initials'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Teams with Most Players")
plt.xlabel("Team")
plt.ylabel("Number of Players records")
plt.xticks(rotation=45)
plt.show()
```

# Observation
The bar chart shows the top 10 teams with the highest number of player records in the dataset. Some teams have significantly more player entries, indicating greater participation across World Cup matches.

# 📈 Visualization 2
Player Position Distribution

```python
plt.figure(figsize=(8,6))

sns.countplot(y='Position', data=df,
              order=df['Position'].value_counts().index)

plt.title("Distribution of Player Positions")
plt.xlabel("Number of Players")
plt.ylabel("Position")

plt.show()
```

### Observation

The bar chart shows the distribution of player positions in the dataset. Goalkeepers (GK) have the highest number of player records, followed by players in the C position. The GKC position has the fewest player records, indicating that it appears less frequently in the dataset.

### Conclusion

The visualization indicates that Goalkeepers (GK) are the most common player position in this dataset. The distribution of player positions is uneven, with some positions appearing much more frequently than others.

## 📊 Visualization 3

### Starting Players vs Substitutes

```python
lineup = df['Line-up'].replace({'S':'Starting Player','N':'Substitute'})

plt.figure(figsize=(6,6))
lineup.value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Starting Players vs Substitutes")
plt.ylabel("")
plt.show()
```

### Observation

The pie chart illustrates the distribution of starting players (S) and non-starting players (N) in the FIFA World Cup Players dataset. The distribution is nearly equal, with **50.4%** of the records belonging to non-starting players (N) and **49.6%** representing starting players (S). This indicates that the dataset contains an almost balanced representation of both categories.

### Conclusion

The visualization shows that the dataset includes a nearly equal proportion of starting players and substitutes/non-starting players. This balanced distribution provides a comprehensive view of player participation in FIFA World Cup matches.

## 📊 Visualization 4

### Top 10 Coaches with Most Player Records

```python
plt.figure(figsize=(12,6))

df['Coach Name'].value_counts().head(10).plot(
    kind='bar',
    color='mediumseagreen'
)

plt.title("Top 10 Coaches with Most Player Records")
plt.xlabel("Coach Name")
plt.ylabel("Number of Player Records")
plt.xticks(rotation=60)

plt.show()
```

### Observation
The bar chart shows the top 10 coaches with the highest number of player records in the dataset. Some coaches are associated with more player records than others, indicating that they managed teams participating in more FIFA World Cup matches.

### Conclusion

The visualization indicates that a few coaches have significantly more player records than others. This reflects their greater involvement in FIFA World Cup tournaments.

## 📊 Visualization 5

### Distribution of Shirt Numbers

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Shirt Number'], bins=25, color='orange')

plt.title("Distribution of Shirt Numbers")
plt.xlabel("Shirt Number")
plt.ylabel("Frequency")

plt.show()
```

### Observation

The histogram shows how shirt numbers are distributed among FIFA World Cup players. Some shirt numbers are used more frequently than others, while higher shirt numbers appear less often.

### Conclusion

The distribution of shirt numbers is uneven, with certain numbers being more commonly assigned to players across different teams.

## 📊 Visualization 6

### Top 10 Match Events

```python
plt.figure(figsize=(12,6))

df['Event'].dropna().value_counts().head(10).plot(
    kind='bar',
    color='purple'
)

plt.title("Top 10 Match Events")
plt.xlabel("Event")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

### Observation

The graph displays the most common match events recorded in the dataset. It highlights the events that occurred most frequently during FIFA World Cup matches.

### Conclusion

The visualization provides insights into the frequency of different match events, helping to understand the most common actions performed by players during matches.

## 📊 Visualization 7

### Missing Values Heatmap

```python
plt.figure(figsize=(10,5))

sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values Heatmap")

plt.show()
```

### Observation

The heatmap displays the locations of missing values in the dataset. Some columns contain more missing data than others, indicating that data cleaning may be required before advanced analysis.

### Conclusion

The dataset contains missing values in a few columns. Identifying these missing values is an important step before performing predictive analysis or machine learning.
